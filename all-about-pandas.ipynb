```python
df.loc[filt]
```

```python
df.loc[filt,'email']
```

```python
df[(df['first']=='sadia') & (df['last']=='sultana')]
```

```python
#we want last name equal to sultana or  1st name equal to samira

filt=(df['last']=='sultana') | (df['first']=='samira')
```

```python
df.loc[filt]
```

```python
# if  we want to oposite the filter just use ~
df.loc[~filt]
```
